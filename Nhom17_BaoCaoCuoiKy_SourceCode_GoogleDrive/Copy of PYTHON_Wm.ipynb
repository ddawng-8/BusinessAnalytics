```python
import numpy as np
import pandas as pd
import statistics as st
import matplotlib.pyplot as plt
from scipy.stats import skew
from scipy.stats import kurtosis
```

```python
data_url = 'https://raw.githubusercontent.com/minknhom/dataset/main/walmart_stock_price.csv'
gdp_lab= pd.read_csv(data_url)
gdp_lab
```

```python
#Hàm lấy số lượng mẫu
len(gdp_lab)
```

```python
#Hàm tìm lớn nhất
min(gdp_lab.Close)
```

```python
#Hàm tìm nhỏ nhất
max(gdp_lab.Close)
```

```python
#Tính trung bình
st.mean(gdp_lab.Close)
```

```python
#Tính Trung Vị
st.median(gdp_lab.Close)
```

```python
#Tính Phương sai
st.variance(gdp_lab.Close)
```

```python
#Tính độ lệnh chuẩn
st.stdev(gdp_lab.Close)
```

```python
#Tính CV
cv = st.stdev(gdp_lab.Close)/st.mean(gdp_lab.Close)
cv
```

```python
skew(gdp_lab.Close)
```

```python
kurtosis(gdp_lab.Close)
```

```python
np.quantile(gdp_lab.Close,0.25)
```
